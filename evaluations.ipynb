```python
from transformers import GPT2Tokenizer, GPT2Model

# Load the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2_dir/')

# Load the model
model = GPT2Model.from_pretrained('gpt2_dir/')

# Set the padding token in the model's configuration
model.config.pad_token_id = tokenizer.pad_token_id

# Save the model
model.save_pretrained('gpt2_dir/')
```
